```
def removing_highly_varying_data_from_one_particular_location(x):
  improved=pd.DataFrame()
  for nameoflocation,subdataframe in x.groupby('location'):
    tempdataframe=pd.DataFrame()
    mean=subdataframe['Price_persqft'].mean()
    std=subdataframe['Price_persqft'].std()
    tempdataframe=subdataframe[(subdataframe['Price_persqft']>=(mean-std)) & (subdataframe['Price_persqft']<=(mean+std))]
    improved=pd.concat([improved,tempdataframe],ignore_index=True)
  return improved 
```

```
def remove_sizeoutliers(x):
  exclude_indices=np.array([])  #creating empty array
  for location,location_df in x.groupby('location'):
    size_stats={} #creating empty dictionary
    for size,size_df in location_df.groupby('size'):
      size_stats[size] ={
                'mean': np.mean(size_df.Price_persqft),
                'std': np.std(size_df.Price_persqft),
                'count': size_df.shape[0]
           
      }
    for size,size_df in location_df.groupby('size'):
      stats=size_stats.get(size-1)
      if (stats and stats['count']>5):
        exclude_indices=np.append(exclude_indices,size_df[size_df.Price_persqft<(stats['mean'])].index.values)
  return x.drop(exclude_indices,axis='index')
dataset10=remove_sizeoutliers(dataset8)         
```
